## Data Pre-processing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('studentscores.csv')
```

```python
#For reference

dataset
```

```python
X = dataset.iloc[ : ,  : 1 ].values
Y = dataset.iloc[ : , 1 ].values

from sklearn.cross_validation import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

## Fitting LR Model to Training Set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor = regressor.fit(X_train, Y_train)
```

## Making Prediction Using Model

```python
Y_pred = regressor.predict(X_test)
```

## Visualization

### Visualizing Training Results

```python
plt.scatter(X_train, Y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')

plt.title("Trained Model Results")
plt.xlabel("Hours")
plt.ylabel("Scores")

plt.show()
```

## Visualizing Test Results

```python
plt.scatter(X_test, Y_test, color = 'red')
plt.plot(X_test, regressor.predict(X_test), color = 'blue')

plt.title("Model Ran Test Results / Predictions")
plt.xlabel("Hours")
plt.ylabel("Scores")

plt.show()
```

```python
## Showing Trained Model Results with Predictions on top to show closeness/difference


#Actual and Predictions
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

p = figure(plot_width=450, plot_height=450, title="L.R.: Actual (blue) & Predictions (red)")


p.line(list(range(10)), Y_train, line_width=2) #Actual

p.line(list(range(10)), Y_test, color="firebrick", line_width=2) #Predictions

p.xaxis.axis_label = "Hours"
p.yaxis.axis_label = "Scores"

show(p)
```

#### Due to lack of overlap between actual and predicted results, seems like the model can be retrained with more data or with model parameters set to generalize more. This is an attempt to have actual results closer to predicted results, which would mean model would be a strong one. And that is the aim to make good predictions.
